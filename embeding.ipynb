```python
from tensorflow.keras.preprocessing.text import  one_hot  
```

```python
# sentences
# The embedding layer converts each word index into a dense vector representation of fixed size.
sent=[" the glass of milk"
     ,'the glass of juice',
     'I am a good boy',
     'understand the meaning of words',
'your videos are good'
    ]
```

```python
sent
```

```python
print(type(sent))
```

```python
#define the vocabulary size
voc_size=10000
```

```python
#one hot representation
one_hot_repr=[one_hot(words,voc_size) for words in sent]
one_hot_repr
```

[5724, 3464, 4103, 4729]=> the index of the words (the glass of milk) ie 5724-> the all other index will zero at that time 

```python
from tensorflow.keras.layers import Embedding
#from tensorflow.keras.preprocessing.sequence import   pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
  
```

```python
import numpy as np
sent_length=8 #max length of the sentence
```

```python
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
embedded_docs#fixed length vector of 8 for each sentence
```

```python
#feature dimension
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
# Each number → converted to a vector of size = dim (10)
```

```python
model.summary()
```

```python
model.predict(embedded_docs)  # converting words → vectors
```
